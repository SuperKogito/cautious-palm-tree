## Constant linear fbanks

```python
from spafe.utils import vis
from spafe.fbanks import linear_fbanks

# init vars
nfilts = 24
nfft = 512
fs = 16000
low_freq = 0
high_freq = 8000
scale = "constant"

# compute the linear filter banks
lin_filbanks = linear_fbanks.linear_filter_banks(nfilts=nfilts,
                                              nfft=nfft,
                                              fs=fs,
                                              low_freq=low_freq,
                                              high_freq=high_freq,
                                              scale=scale)
# plot filter banks
vis.visualize_fbanks(lin_filbanks, "Amplitude", "Frequency (Hz)")
```
